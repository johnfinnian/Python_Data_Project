The copy() method makes a copy of this object's indices and data.
TO use this method, we have to create a new dataframe

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

Why should copy be used as against a new varible of a given dataset?  With the use of new variable, changes made can affect the original data frame. Which is proven with examples and id() function below.

```python

df_altered =df_original

```

```python
df_altered.loc[:5, 'salary_year_avg']
```

```python
median_salary = df_altered['salary_year_avg'].median()

median_salary
```

```python
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)
```

```python
df_altered.loc[:5, 'salary_year_avg']
```

```python
df_original.loc[:5, 'salary_year_avg']
```

```python
id(df_original)
```

```python
id(df_altered)
```

```python
id(df_original) == id(df_altered)
```

This can be avoided by using copy() method when creating a variable. As shown below

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
df_altered = df_original.copy()
```

```python
print('ID of df_original:', id(df_original))
print('ID of df_altered:', id(df_altered))
print('Are the IDs of the two DataFrames equal?', id(df_original) == id(df_altered))
```

```python
median_salary = df_altered['salary_year_avg'].median()

df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)

df_altered.loc[:5, 'salary_year_avg']
```

```python
df_original.loc[:5, 'salary_year_avg']
```

Therefore the use of copy method protects the original dataframe.
