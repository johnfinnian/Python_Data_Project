```python
# Install datasets Library
# !pip install datasts

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.loc[0:10, 'salary_rate':'salary_hour_avg']
```

```python
median_year_avg = df['salary_year_avg'].median()
```

```python
median_hour_avg = df['salary_hour_avg'].median()
```

```python
df_filled = df
```

```python
df_filled['salary_year_avg'].fillna(median_year_avg)
```

```python
df_filled['salary_year_avg']
```

```python
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_year_avg)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_hour_avg)
```

here we fill in the null values in salary year avg with median salary using fillna() method.

```python
df_filled['salary_year_avg']
```

```python
df_filled['salary_hour_avg']
```

Haven filled all the null values in salary with the median salaries, the next thing is to drop duplicates. This is done with a pandas method in data frame; drop_duplicates()

```python
df_unique = df_filled

df_unique = df_unique.drop_duplicates()
```

```python
print('Length of original df:', len(df_filled))
print('Length of drop duplicates df:', len(df_unique) )
print('Rows dropped:', len(df_filled) - len(df_unique))
```

With drop_duplicate() method, we will further drop any duplicates in two specific columns; job_title and company name.

```python
df_unique = df_unique.drop_duplicates(subset=['job_title', 'company_name'])

print('Length of original df:', len(df_filled))
print('Length of drop duplicates df:', len(df_unique) )
print('Rows dropped:', len(df_filled) - len(df_unique))
```
