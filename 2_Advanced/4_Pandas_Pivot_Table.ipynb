```python
# Install datasets Library
# !pip install datasts

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

COUNT OF JOBS TITLES

```python
df.pivot_table(index='job_title_short', aggfunc='size')

# size attribute counts all rows in a group(Value/Column not required)
# count attribute counts non-null values in a group(Value/Column required)
```

```python
df.groupby('job_title_short').size()
```

MEDIAN SALARY by JOB

```python
df.pivot_table(values='salary_year_avg',index='job_title_short', aggfunc='median')
```

```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

MEDIAN SALARY ANALYSIS by COUNTRY

```python
df.pivot_table(values='salary_year_avg',index='job_title_short', columns='job_country',aggfunc='median')
```

```python
df.pivot_table(values='salary_year_avg',index='job_country', columns='job_title_short',aggfunc='median')
```

```python
df['job_country'].value_counts().head(6).index
```

```python
top_countries =df['job_country'].value_counts().head(6).index

df_job_country_salary = df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = ['Data Scientist', 'Data Engineer', 'Data Analyst']

df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary (USD)')
plt.xlabel('')
plt.title('Median Salary by Job Country and Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```
